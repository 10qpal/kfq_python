# 데이터 전처리

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cust_df = pd.read_csv('./train_santander.csv')
cust_df
```

```python
cust_df.info()
```

```python
cust_df['TARGET'].value_counts()
```

```python
cust_df.describe()
```

```python
# var3 피처에서 -999999 값을 2로 변환하기(inplace : 변환한 내용 바로 적용하기)
# ID 피처 지우기
cust_df['var3'].replace(-999999, 2, inplace=True)
cust_df.drop('ID', axis=1, inplace=True)
```

```python
# 데이터(X_features), 타겟(y_labels) 분리
X_features = cust_df.iloc[:,:-1]
y_labels = cust_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size=0.2, random_state=0)
```

```python
# 학습 세트, 테스트 세트 레이블(타겟) 값 분포 비율 확인
train_cnt = y_train.count()
test_cnt = y_test.count()

print("학습 세트 레이블 값 분포 비율")
print(y_train.value_counts()/train_cnt)
print()
print("테스트 세트 레이블 값 분포 비율")
print(y_test.value_counts()/test_cnt)
```

# XGBoost 모델 학습과 하이퍼 파라미터 튜닝

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# n_estimators : 500
xgb_clf = XGBClassifier(n_estimators=500, random_state=156)

# 성능 평가 지표 : auc, 조기 중단 파라미터 : 100
xgb_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc',
           eval_set=[(X_train, y_train), (X_test, y_test)])
```

```python
xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1], average='macro')
xgb_roc_score
```

```python
from sklearn.model_selection import GridSearchCV

# n_estimators : 100
xgb_clf = XGBClassifier(n_estimators=100)
params = {'max_depth' : [5, 7],
         'min_child_weight' : [1, 3],
         'colsample_bytree' : [0.5, 0.75]
         }

# cv : 3
grid_cv = GridSearchCV(xgb_clf, param_grid=params, cv=3)
grid_cv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric='auc',
           eval_set=[(X_train, y_train), (X_test, y_test)])
```

```python
print(grid_cv.best_params_)
print(grid_cv.best_score_)

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1], average='macro')
print(xgb_roc_score)
```

# LightGBM 모델 학습과 하이퍼 파라미터 튜닝

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
```

```python
lgbm_clf = LGBMClassifier(n_estimators=500)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc',
           eval_set=evals, verbose=True)
```

```python
lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:,1], average='macro')
lgbm_roc_score
```

```python
from sklearn.model_selection import GridSearchCV

lgbm_clf = LGBMClassifier(n_estimators=200)
params = {'num_leaves' : [32, 64],
          'max_depth' : [128, 160],
          'min_child_samples' : [60, 0.100],
          'subsample': [0.8, 1]
         }

grid_cv = GridSearchCV(lgbm_clf, param_grid=params, cv=3)
grid_cv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric='auc',
           eval_set=[(X_train, y_train), (X_test, y_test)])
```

```python
grid_cv.best_params_
```

```python
grid_cv.best_score_
```
