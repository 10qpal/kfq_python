```python
import folium
import pandas as pd
import requests
import json
import webbrowser
```

```python
filename = 'jeju.csv'
df = pd.read_csv(filename, encoding='cp949', header = 0)
df
```

```python
def net_get_geo_coder(query):
    KAKAO_API_KEY = '5e66881ea3128e3ee45eb7160a210c2d'
    url = '''
    https://dapi.kakao.com/v2/local/search/keyword.json?query={0}&category_group_code=PO3
    '''.format(query)
    headers = {'Authorization':'KakaoAK {0}'.format(KAKAO_API_KEY)}
    
    # API 쿼리 수행
    res = requests.get(url, headers = headers)
    res = res.json()
    try:
        return[float(res['documents'][0]['y']), float(res['documents'][0]['x'])]
    except:
        return '결과 없음'
```

```python
result = net_get_geo_coder('제주특별자치도 제주시 만덕로 18')
result
```

```python
df['주소'] = df['주소'].str.split('(').str.get(0)
df.head(3)
```

```python
import folium
map = folium.Map(location=[33.3616658, 126.5204118], zoom_start = 10)
for index, row in df.iterrows():
    print(index)
    geoData = net_get_geo_coder(row['주소'])
    print(geoData)
    if geoData != None and geoData != '결과 없음':
        folium.Marker(geoData, popup=row['읍면동'], icon=folium.Icon(color='red')).add_to(map)
map
```

```python
svFilename = 'jeju.html'
map.save(svFilename)

import webbrowser
webbrowser.open(svFilename)
```
