### 산포도

```python
data = {'a' : np.arange(50),
       'c' : np.arange(50),
       'd' : np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
plt.scatter('a', 'b', c='c', s='d', marker='s', data=data)    # x, y, color, size
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.minorticks_on()    # 보조눈금선
plt.show()
```

### 한 번에 여러 그래프 그리기

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.figure(figsize=(9, 3))    # 가로 9, 세로 3
plt.subplot(131)    # 1행 3열 중 1번째 위치
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### 히스토그램

```python
# 히스토그램 데이터
mu,sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)    # 밀도, 색상, 투명도
# print(n)
# print(bins)
# print(patches)
plt.xlabel('Smarts', fontsize=14, color='red')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100, \sigma=15$')    # 위치, 텍스트
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### annotate

```python
ax = plt.subplot(111)
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
# lw : 선 굵기
line = plt.plot(t, s, lw = 2)
# xy : x, y 위치 / xytext : 텍스트 위치 / arrowprops : dict(화살표 색상, 화살표 길이)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.ylim(-2, 2)
plt.show()
```

### 막대그래프

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # x 위치
width = 0.35    # 막대 너비
plt.rcParams['font.family'] = 'NanumBarunGothic'
plt.rcParams['font.size'] = 14
p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans, yerr=menStd)    # 위, 아래에 나타내기 위해 bottom 설정
plt.ylabel('점수')
plt.xlabel('그룹')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
# 범례
plt.legend((p1[0], p2[0]), ('남자', '여자'))
```
