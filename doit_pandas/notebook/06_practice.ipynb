# 누락값 확인하기

```python
from numpy import NaN, NAN, nan
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
```

```python
print(NaN == NaN)
print(NaN == nan)
print(NaN == NAN)
```

```python
import pandas as pd
print(pd.isnull(NaN))
print(pd.isnull(NAN))
print(pd.isnull(nan))
```

```python
print(pd.notnull(NaN))
print(pd.notnull(42))
print(pd.notnull('missing'))
```

# 누락값을 포함한 데이터를 불러올 때

```python
visited = pd.read_csv('../data/survey_visited.csv')
survey = pd.read_csv('../data/survey_survey.csv')
print(visited)
print()
print(survey)
```

# 데이터 집합을 연결할 때 누락값이 발생하는 경우

```python
vs = visited.merge(survey, left_on = 'ident', right_on = 'taken')
print(vs)
```

# 누락값의 개수 구하기

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
print(ebola.count())
```

```python
print(ebola.info())
```

```python
import numpy as np
print(np.count_nonzero(ebola.isnull()))
```

```python
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
print(ebola.Cases_Guinea.value_counts().head())
print(ebola.Cases_Guinea.value_counts(dropna = False).head())    # NaN 포함
```
