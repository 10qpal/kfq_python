```python
import pandas as pd
import numpy as np
# test = pd.read_csv('test.csv', encoding='utf-8')
df = pd.DataFrame(columns = ['년도','태풍번호'])
```

```python
file = pd.read_csv('모든태풍.csv',encoding='cp949')
file_length = len(file)

for i in range(file_length):
    if(file['년도'][i] == '년도'):
        file = file.drop(i)
file.reset_index(drop=True, inplace=True)

tp_no = 0
count = 0
col_i = 0
```

```python
print(file['위도'][0][:-1])
print(file['경도'][0][:-1])
```

```python
for i in range(0, file_length-1):
    count+=1
    x = file['위도'][i][:-1]
    y = file['경도'][i][:-1]
    df.loc[tp_no,'x'+str(count)] = x
    df.loc[tp_no,'y'+str(count)] = y
    col_i = count*2
    if(file.iloc[i][0] != file.iloc[i+1][0] or file.iloc[i][1] != file.iloc[i+1][1]):
        df.loc[tp_no,'년도'] = file.iloc[i][0]    #년도
        df.loc[tp_no,'태풍번호'] = file.iloc[i][1]    #년도의 태풍번호
#         for j in range(count*2, len(df.columns)-1, 2):
#             df.iloc[tp_no,col_i+2] = x
#             df.iloc[tp_no,col_i+3] = y
        tp_no += 1
        count = 0
```

```python
tp_no = 0
count = 0
col_i = 0
for i in range(0, file_length-1):
    count+=1
    col_i = count*2
    if(file.iloc[i][0] != file.iloc[i+1][0] or file.iloc[i][1] != file.iloc[i+1][1]):
        for j in range(col_i+2, len(df.columns), 2):
#             df.iloc[tp_no][col_i+2].replace(np.nan, df.iloc[tp_no][col_i])
#             df.iloc[tp_no][col_i+3].replace(np.nan, df.iloc[tp_no][col_i+1])
            df.iloc[tp_no][col_i+2].replace(np.nan, '0')
            df.iloc[tp_no][col_i+3].replace(np.nan, '0')
        tp_no += 1
        count = 0
```
